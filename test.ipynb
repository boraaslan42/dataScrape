```python
import requests
import xml.etree.ElementTree as ET

def parse_xml_from_web(url):
    response = requests.get(url)  # Send a GET request to the URL
    xml_content = response.content  # Get the XML content from the response
    
    # Parse the XML content
    root = ET.fromstring(xml_content)
    
    # Process the XML data
    # Example: Print the tag name of each element in the XML
    for element in root.iter():
        print(element)

# Example usage
url = "https://www.pngmart.com/post-sitemap4.xml"  # Replace with the URL of the XML page
parse_xml_from_web(url)
```
